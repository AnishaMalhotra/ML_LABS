```python
from sklearn import datasets
faces = datasets.fetch_olivetti_faces()
faces.data.shape
```

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(8, 6))
# plot several images
for i in range(15):
    ax = fig.add_subplot(3, 5, i + 1, xticks=[], yticks=[])
    ax.imshow(faces.images[i], cmap=plt.cm.bone)
```

```python
#WITH PCA

from sklearn import decomposition

# reducing features using PCA
pca = decomposition.PCA(n_components=150)

X_std_pca = pca.fit_transform(faces.data)
print(X_std_pca.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std_pca,
faces.target, random_state=0)
print(X_train.shape, X_test.shape)
```

```python
plt.imshow(pca.mean_.reshape(faces.images[0].shape),cmap=plt.cm.bone)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score

y_pred = gnb.predict(X_test)

precision = precision_score(y_test, y_pred, average="micro")
print("precision: ", precision)

recall = recall_score(y_test, y_pred, average="micro")
print("recall: ", recall)
```
